```python
from pymatgen.ext.matproj import MPRester
from pymatgen.phonon.bandstructure import *
import numpy as np
from pymatgen.phonon.plotter import*
with MPRester("uI8z9nwQlS8xKKpAxb") as m:
    phonon_bs = m.get_phonon_bandstructure_by_material_id("mp-10378")

plotter = PhononBSPlotter(phonon_bs)
plotter.get_plot()
plotter.plot_brillouin()
```

Les bandes accoustiques se trouvent en regardant les bandes où la fréquence du phonon associé est nulle. Ici se sont $\Gamma-X$, $K-\Gamma$ et $\Gamma-L$

```python
data_phonon_bs = plotter.bs_plot_data()
distance_phonon_bs  = data_phonon_bs.get("distances")
frequence_phonon_bs = data_phonon_bs.get("frequency")

vitesse_1 = 2 * np.pi * abs((frequence_phonon_bs[0][0][1]-frequence_phonon_bs[0][0][0]
                            )*(10**12))/((distance_phonon_bs[0][1]-distance_phonon_bs[0][0])*(10**10))

vitesse_2 = 2 * np.pi * abs((frequence_phonon_bs[3][1][-1]-frequence_phonon_bs[3][1][-2]
                            )*(10**12))/((distance_phonon_bs[3][-1]-distance_phonon_bs[3][-2])*(10**10))

vitesse_3 = 2 * np.pi * abs((frequence_phonon_bs[4][2][1]-frequence_phonon_bs[4][2][0]
                            )*(10**12))/((distance_phonon_bs[4][1]-distance_phonon_bs[4][0])*(10**10))

print("La vitesse du son de la courbe 1 dans la direction \u0393 - X vaut :", vitesse_1, "m/s")
print()
print("La vitesse du son de la courbe 2 dans la direction K - \u0393 vaut :", vitesse_2, "m/s")
print()
print("La vitesse du son de la courbe 3 dans la direction \u0393 - L vaut :", vitesse_3, "m/s")
```
