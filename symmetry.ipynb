```python
from pymatgen.ext.matproj import MPRester
from pymatgen.core.operations import SymmOp
from pymatgen.core.structure import Structure
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
from IPython.display import Image

with MPRester("uI8z9nwQlS8xKKpAxb") as m:
    structure = m.get_structure_by_material_id("mp-10378")
coords = structure.frac_coords
print(coords)
```

```python
sym_op1 = SymmOp.from_xyz_string("-x, -y, -z") #opération de symétrie pour l'inversion
print("Les coordonées initiales sont :", coords[1])
print("La matrice de rotation est :")
print(sym_op1.rotation_matrix)
print("Le vecteur de translation est :", sym_op1.translation_vector)
new_atom_position1 =sym_op1.operate(coords[1])
print("Les coordonées finales sont :", new_atom_position1)
Image(filename="inverse.PNG")
```

```python
sym_op2 = SymmOp.from_xyz_string("-y, x, z") #opération de symétrie pour l'inversion
print("Les coordonées initiales sont :", coords[0])
print("La matrice de rotation est :")
print(sym_op2.rotation_matrix)
print("Le vecteur de translation est :", sym_op2.translation_vector)
new_atom_position2 =sym_op2.operate(coords[0])
print("Les coordonées finales sont :", new_atom_position2)
Image(filename="C4.PNG")
```

```python
sym_op3 = SymmOp.from_xyz_string("x+1/2,y, z+1/2") #opération de symétrie pour l'inversion
print("Les coordonées initiales sont :", coords[2])
print("La matrice de rotation est :")
print(sym_op3.rotation_matrix)
print("Le vecteur de translation est :", sym_op3.translation_vector)
new_atom_position3 =sym_op3.operate(coords[2])
print("Les coordonées finales sont :", new_atom_position3)
Image(filename="translation.PNG")
```
